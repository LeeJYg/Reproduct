```python
import matplotlib.pyplot as plt

# 가정: 'training_log.txt' 파일에 로그가 저장되어 있음
file_path = '/home/jylee/Reproduct/week1_repro/training_loss_records.txt'

# 에포크와 손실 값을 저장할 리스트
epochs = []
losses = []

# 파일을 열고 내용을 읽음
with open(file_path, 'r') as file:
    for line in file:
        # 'Epoch'과 'Loss'가 포함된 줄에서 값을 추출
        if 'Epoch' in line and 'Loss' in line:
            parts = line.split(',')
            # 'Epoch:'과 'Loss:' 문자열을 제거하고 값을 정수/실수로 변환
            epoch = int(parts[0].split(':')[1].strip())
            loss = float(parts[1].split(':')[1].strip())
            epochs.append(epoch)
            losses.append(loss)

train_epochs = []
train_losses = []
eval_epochs = []
eval_losses = []

for i, line in enumerate(zip(epochs, losses)):
    epoch = line[0]
    loss = line[1]
    
    if i % 2 == 0:  # Even line, training loss
        train_epochs.append(epoch)
        train_losses.append(loss)
    else:  # Odd line, evaluation loss
        eval_epochs.append(epoch)
        eval_losses.append(loss)

# Plotting
plt.figure(figsize=(14, 7))


plt.plot(train_epochs, train_losses, 'b-o', label='Training Loss')

#plt.plot(eval_epochs, eval_losses, 'r--s', label='Evaluation Loss')

plt.title('Training and Evaluation Loss over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```
